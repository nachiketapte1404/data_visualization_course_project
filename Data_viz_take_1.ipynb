```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('original_dataset.csv',encoding='latin-1')
```

```python
df.head()
```

```python
df = df.drop(columns=['DisNo.','Historic','Classification Key','Disaster Group','External IDs','Event Name','ISO','Origin','Associated Types','Latitude','Longitude','Admin Units','Entry Date','Last Update'],axis=1)
```

```python
df.head()
```

```python
df[["AID Contribution ('000 US$)","No. Affected", "No. Homeless", "Reconstruction Costs ('000 US$)", "Reconstruction Costs, Adjusted ('000 US$)","Insured Damage ('000 US$)","Insured Damage, Adjusted ('000 US$)","Total Damage ('000 US$)","Total Damage, Adjusted ('000 US$)"]] = df[["AID Contribution ('000 US$)","No. Affected", "No. Homeless", "Reconstruction Costs ('000 US$)", "Reconstruction Costs, Adjusted ('000 US$)","Insured Damage ('000 US$)","Insured Damage, Adjusted ('000 US$)","Total Damage ('000 US$)","Total Damage, Adjusted ('000 US$)"]].fillna(0)
```

```python
df.isna().sum()
```

```python
# Column to check
columns_to_check = ['Total Deaths','Start Day','End Day','Location']
for column_to_check in columns_to_check:
  # Remove rows where the value in the specified column is 0 or NA
  df = df[~df[column_to_check].isna() & (df[column_to_check] != 0)]

df.shape
```

```python
df.isna().sum()
```

```python
df['Start Year'] = df['Start Year'].astype(int)
df['Start Month'] = df['Start Month'].astype(int)
df['Start Day'] = df['Start Day'].astype(int)
df['End Year'] = df['End Year'].astype(int)
df['End Month'] = df['End Month'].astype(int)
df['End Day'] = df['End Day'].astype(int)
df['Start_Date'] = pd.to_datetime(df['Start Year'].astype(str) + '-' + df['Start Month'].astype(str) + '-' + df['Start Day'].astype(str))
df['End_Date'] = pd.to_datetime(df['End Year'].astype(str) + '-' + df['End Month'].astype(str) + '-' + df['End Day'].astype(str))
df['Duration'] = (df['End_Date'] - df['Start_Date']).dt.days
df
```

```python
df = df.drop(columns=['Start Year','Start Month','Start Day','End Day','End Month','End Year'],axis=1)
df.head()
```

```python
df.isna().sum()
```

```python
df['CPI'] = df['CPI'].fillna(0)
df.isna().sum()
```

```python
df.shape
```

```python
df.to_csv('cleaned_data.csv', index=False)
```
